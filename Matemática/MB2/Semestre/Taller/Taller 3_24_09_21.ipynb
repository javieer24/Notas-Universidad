# Simplificación Algebraica
Algunos de los comando de simplificación son los siguientes:

* factor() "factorizar"
* expand() "multiplican expresiones"
* simplify() "Simplifican expresiones"
* collect() "Reunir"
* apart() "fracciones parciales" 
* subs() "sustituir expresiones"

```python
factor(x**2+5*x+6)
expand((x**2+3)*(x**3+1))
simplify(cos(x)**2+sin(x)**2)
collect(expand((x+1)**2*(y+1)**3),y)
apart(x/(x**2+5*x+6))
expr=apart(x/(x**2+5*x+6))
expr.subs(x,3) #valor a sustituir
```

## Ecuaciones
Para resolver ecuaciones primero defines con Eq() y luego se utiliza el comando solve  para obtener la soluciòn

```python

e=Eq(x**2+5*x+6,cos(x))
solve(x**2+5*x+5)# cuadrática
solve([x+y-1,x-y],[x,y])# sistema lineal
solve([x**2+y**2-25,x-y])# sistema no lineal
# otro
r,r1,r2=symbols("r r1 r2")
e=Eq(1/r,1/r1+1/r2)
e
solve(e,r)  
```

# Limites, Sumatorias, Derivadas,  integrales y series

* limit() #limites
* summation() #sumatorias
* diff() #derivadas
* integrate() #integrales
* series() #series

```python
limit(sin(x)/x,x,0)  # límite  lateral
limit((x**2-1)/(x+1),x,-1) # límite de función
limit((3*x**2+5*x+6)/(4*x**2-3*x+4),x,oo) # límite al infinito
a=symbols("a",negative=True)
limit(1/(1+exp(a*x)),x,-oo) # otro ejemplo
```

```python
summation(1/x,(x,1,10)).n()
```

```python
diff(cos(x)) #derivada
diff(exp(2*x),(x,10))   #derivada de orden superior
(x**2+5*x+6).diff()
diff(x**2+y**2+5*x*y-2,y)
(x**2+y**2+5*x*y-2).diff(y,2) 
```

```python
integrate(a*x**2+3,x) #integrales
integrate(cos(x),(x,1,4)).n() #integrales definidas
series(1/(3+x),x,0,10) #serie de taylor 1variable,2centrada, 3iteraciones
```

## Geometria
* Puntos --> Point()
* lineas --> Line()
* Circulos--> Circle()
* triangulos -->Triangle()
* area -->  objeto.area
* interseccion -->interseccion(a,b) 

```python
p1=Point(0,0)  #punto (0,0)
p2=Point(a,0)  #punto  (a,0)
p3=Point(a,b)
t=Triangle(p1,p2,p3)  # triángulo con puntos p1,p2,p3
t.area                   # área de un triángulo
c=Circle(p1,10)
l=Line(Point(10,10),Point(10,-10))  # linea
c.is_tangent(l)                     # prueba de tangencia
l2=Line(Point(1,1),Point(1,-1))
intersection(c,l2)                  #intersección de circulo y linea

l3=Line(Point(0,3),Point(6,0))   
c2=Circle(Point(2,2),2)
intersection(c2,l3)
```

# Vectorización de funciones  
* vectorize() - función
* lambdify()  - expresion de numpy


```python
import numpy as np
# defininción de funcion factorial
def factorial(n):
    if n==0:
        return(1)
    else:
        return(n*factorial(n-1))
factorial_np=np.vectorize(factorial)  # vectorización de funcion propia
factorial_np(np.array([1,2,3,4,5,6,7]))
expr=exp(x)*cos(x)
f=lambdify(x,expr,"numpy")# expresión de SymPy    
f(np.linspace(1,20,6))   # función  valida para arreglo de numpy
expr2=exp(x)*cos(y)
f2=lambdify((x,y),expr2,"numpy")
f2(np.array([1,5,6]),np.array([4,6,7]))
```
